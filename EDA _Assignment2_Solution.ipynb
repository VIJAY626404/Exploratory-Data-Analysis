# 1. Identify the top 5 popular artists based on the mean popularity of their tracks.Show the mean popularity of tracks for the top 5 popular artists Using BarPlot 

```python
spotify_data['Artist'].dtype
```

```python
spotify_data['Popularity'].dtype
```

```python
d1 = spotify_data.groupby(['Artist'])['Popularity'].mean()
d1= pd.DataFrame(d1)
d1 = d1.reset_index().sort_values(by='Popularity',ascending=False)
d1 = d1.rename(columns={'Popularity':'Mean Popularity'})
d1= d1.reset_index(drop = True)
```

```python
top_five = d1.head(5)
```

```python
top_five
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top_five,y='Mean Popularity',x='Artist')
plt.title("Top 5 Populer Artist")
plt.xlabel("Top Artist")
plt.ylabel("Mean Popularity")
plt.show()
```

# 2. Determine the top 5 popular songs based on their popularity ratings.Display the popularity ratings of the top 5 popular songs using BarPlot.

```python
p = spotify_data.groupby(['Track Name'])['Popularity'].max().reset_index()

# Rename the columns and sort by 'Popularity' in descending order
p = p.rename(columns={'Track Name': 'Populer Songs'}).sort_values(by='Popularity', ascending=False)

# Reset index without creating a new index column
p = p.reset_index(drop=True)
```

```python
populer_songs = p.head(5)
populer_songs
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=populer_songs,x='Populer Songs',y = 'Popularity')
plt.title("Top 5 Populer Songs based on Popularity rating")
plt.xlabel("Top Songs")
plt.ylabel("Popularity")
plt.show()
```

# 3. Find the top 5 trending genres based on the mean popularity of tracks within each genre.Visualize the mean popularity of tracks for the top 5 trending genres.

In Spotify Dataset genres related column are not given that's why i am using Track Name and Popularity column for finding top 5 treding genres

```python
g = spotify_data.groupby(['Track Name'])['Popularity'].mean().reset_index()

# Rename the columns and sort by 'Popularity' in descending order
g = g.rename(columns={'Track Name': 'Trending Genres','Popularity':'Mean Popularity'})
g =g.sort_values(by='Mean Popularity', ascending=False)

# Reset index without creating a new index column
g = g.reset_index(drop=True)
```

```python
trending_genres = g.head()
trending_genres
```

```python
plt.figure()
sns.lineplot(data=trending_genres,y = 'Trending Genres',x = 'Mean Popularity',marker='o', color='red')
plt.title('Trend of Mean Popularity Across Top 5 Trending Genres')
plt.xlabel('Genres')
plt.ylabel('Mean Popularity')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
top_longest_song = s.head(5)
top_longest_song
# top 5 longest song among the track of the top 5 popular artist are below:
```

# 5. Determine the top 5 most danceable songs based on their danceability scores.Illustrate the danceability scores of the top 5 most danceable songs using PieChart

```python
d = spotify_data.groupby(['Track Name'])['Popularity'].max().reset_index()
d = pd.DataFrame(d)
d = d.sort_values(by='Popularity',ascending=False)
d = d.rename(columns={'Track Name':'Danceable Songs','Popularity':'Danceability Score'})
d = d.reset_index(drop=True)
```

```python
danceable_songs = d.head(5)
danceable_songs
```

```python
danceable_songs['Danceability Score'].dtype
```

```python
plt.figure()
plt.pie(x=danceable_songs['Danceability Score'],autopct="%1.1f%%",labels = danceable_songs['Danceable Songs'],startangle=140,explode=(0.01,0.01,0.01,0.01,0.01))
plt.title('danceability scores of the top 5 most danceable songs')
plt.show()
```

# Insights of Above plot:
1. Highest Danceability Score is 20.5% 
2. Populer Danceable songs is Lovin On Me
